```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data/train7(all_numerical).csv')
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['ni.age','len.at.res','premium','ni.marital.status','n.adults','n.children','tenure','claim.ind','ni.gender.lower','sales.channel.lower','coverage.type.lower','dwelling.type.lower','credit.lower']],data['cancel'], test_size=0.1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn import linear_model
mymodel = linear_model.LogisticRegression(max_iter=1000000)
mymodel.fit(X_train,y_train)
```

```python
mymodel.predict(X_test)
```

```python
X_test
```

```python
mymodel.score(X_test, y_test)
```

```python
predicted_output = mymodel.predict(X_test)
predicted_output
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_output)
cm
```

```python
import seaborn as sn
from matplotlib import pyplot as plt
plt.figure(figsize = (5,4))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted Value')
plt.ylabel('Truth or Actual Value')
```
